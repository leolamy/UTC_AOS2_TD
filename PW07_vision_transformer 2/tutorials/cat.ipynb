# Quick: `torch.cat`

## `torch.cat`

`torch.cat` **concatenates tensors along a specified dimension**.

``` python
torch.cat(tensors, dim=0)
```

-   `tensors`: a sequence (list/tuple) of tensors
-   `dim`: the axis along which to concatenate
-   All tensors must have the same shape *except* for the concatenation
    dimension.

Common uses include combining batches, stacking feature maps, and
merging sequences.

## Example

````python
import torch

a = torch.tensor([
    [1, 2],
    [3, 4]
])  # shape (2, 2)

b = torch.tensor([
    [5, 6],
    [7, 8]
])  # shape (2, 2)

# Concatenate along dim=0 → shape (4, 2)
out0 = torch.cat([a, b], dim=0)

# Concatenate along dim=1 → shape (2, 4)
out1 = torch.cat([a, b], dim=1)

print("Concat dim=0:\n", out0)
print("Concat dim=1:\n", out1)
````
