# Quick: `torch.transpose`

## `torch.transpose`

`torch.transpose` **swaps two dimensions** of a tensor.

``` python
torch.transpose(tensor, dim0, dim1)
```

-   `dim0`: the first dimension to swap
-   `dim1`: the second dimension to swap
-   Returns a **view** with the two axes exchanged.

Common uses include switching between channel-first and channel-last
formats, or rearranging tensor axes before operations like matrix
multiplication.

## Example

import torch

x = torch.tensor(\[ \[1, 2, 3\], \[4, 5, 6\]\]) \# shape (2, 3)

# Swap dims 0 and 1 → shape becomes (3, 2)

y = torch.transpose(x, 0, 1)

print(“Original shape:”, x.shape) print(“Transposed shape:”, y.shape)
print(y)
