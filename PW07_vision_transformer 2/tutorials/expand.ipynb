# Quick: `torch.expand`

## torch.Tensor.expand

`expand()` makes a tensor *look bigger* by repeating values **without
copying memory**. The expanded tensor is a **view**: multiple elements
reference the *same underlying data*.

Use it for cheap broadcasting or duplicating parameters without
allocating new tensors.

## Example

import torch

x = torch.tensor(\[10, 20\]) \# shape (2,) y = x.expand(3, 2) \# pretend
it’s (3,2)

print(“y:”, y)

y\[0, 0\] = 11 y\[1, 1\] = 21 print(“y:”, y)
