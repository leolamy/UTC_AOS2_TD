# Quick: `torch.bmm`

## `torch.bmm`

`torch.bmm` performs **batch matrix multiplication**. It multiplies two
batches of matrices:

-   `A`: (batch, m, n)
-   `B`: (batch, n, p)
-   Result: (batch, m, p)

Each pair of matrices in the batch is multiplied independently.

## Example

```python
import torch

A = torch.tensor([
    [[1., 2.],
     [3., 4.]],
    [[5., 6.],
     [7., 8.]],
])  # shape (2, 2, 2)

B = torch.tensor([
    [[10., 20.],
     [30., 40.]],
    [[1.,  2.],
     [3.,  4.]],
])  # shape (2, 2, 2)

out = torch.bmm(A, B)
print(out)
```
