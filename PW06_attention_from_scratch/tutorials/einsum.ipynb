# Quick: `torch.einsum`

## `torch.einsum`

`torch.einsum` performs tensor operations using **Einstein summation
notation**. It lets you express many operations—matrix multiplication,
reductions, transposes, outer products—in a single compact string.

General form:

``` python
torch.einsum("subscripts", operands...)
```

Examples of common patterns:

-   Matrix multiply: `"ij,jk->ik"`
-   Batch matmul: `"bij,bjk->bik"`
-   Transpose: `"ij->ji"`
-   Sum over an axis: `"ij->i"`

## Example

````python
import torch

a = torch.tensor([1, 2, 3])
b = torch.tensor([10, 20])

out = torch.einsum("i,j->ij", a, b)
print(out)

print(torch.outer(a, b))
````
